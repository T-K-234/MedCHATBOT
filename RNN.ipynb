```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense


```

```python
# Load dataset from CSV
data = pd.read_csv("shah-d.csv")
```

```python
# Separate features and labels
X = data['Symptoms']
y = data['Disease']

```

```python
# Tokenize the text data
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X)
X_sequences = tokenizer.texts_to_sequences(X)

# Pad sequences to ensure uniform length
max_sequence_length = max([len(seq) for seq in X_sequences])
X_padded = pad_sequences(X_sequences, maxlen=max_sequence_length, padding='post')

# Convert labels to categorical format
num_classes = len(set(y))
y_categorical = pd.get_dummies(y).values

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_padded, y_categorical, test_size=0.25, random_state=42)

# Build the RNN model
embedding_dim = 50
model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=embedding_dim, input_length=max_sequence_length))
model.add(LSTM(100))
model.add(Dense(num_classes, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=64, validation_split=0.2)

```

```python
# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```

```python
# Assuming 'model' is already trained and 'tokenizer' is already fitted

# Input text for classification
input_text = [
    "bone spurs have developed around my joint",
]

# Tokenize and pad the input text
input_sequences = tokenizer.texts_to_sequences(input_text)
input_sequences_padded = pad_sequences(input_sequences, maxlen=max_sequence_length, padding='post')

# Predict the class of the input text
predictions = model.predict(input_sequences_padded)

# Get the predicted classes
predicted_classes = np.argmax(predictions, axis=1)

# Map predicted classes to disease names
predicted_diseases = [list(set(y))[pred_class] for pred_class in predicted_classes]

# Print the predicted diseases for the input text
for text, predicted_disease in zip(input_text, predicted_diseases):
    print(f"Text: {text}")
    print(f"Predicted Disease: {predicted_disease}")
    print()
```
