```python
import pandas as pd
from fuzzywuzzy import fuzz
import pickle

# Step 1: Load the dataset from Excel
data = pd.read_excel('final_llm (1).xlsx')

# Step 2: Define inference text and minimum similarity threshold
inference_text = "big fan bro."
min_similarity = 80  # Adjust this threshold as needed (0-100)

# Step 3: Define fuzzy matching function
def fuzzy_match(inference_text, data, min_similarity):
    matches = []
    for text in data['Inputs']:
        similarity = fuzz.ratio(inference_text, text)  # Ratio metric for overall similarity
        if similarity >= min_similarity:
            matches.append(text)
    return matches

# Step 4: Determine the prediction
matches = fuzzy_match(inference_text, data, min_similarity)
prediction = 1 if matches else 0

# Step 5: Save the fuzzy matching function
with open('fuzzy_matching_function.pkl', 'wb') as f:
    pickle.dump(fuzzy_match, f)

# Step 6: Print the prediction
print("Prediction:", prediction)
```
